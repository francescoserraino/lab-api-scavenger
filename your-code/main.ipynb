```python
response = requests.get('https://api.github.com/repos/ta-data-lis/lab-mysql-first-queries/commits')
```

```python
response.json()
```

```python
sec_dataframe = pd.DataFrame(response.json())

commits = pd.DataFrame(dict(sec_dataframe['commit'])).T
commits_more = pd.DataFrame(dict(commits['author'])).T
commits_more
```

```python
import datetime
from datetime import datetime, timedelta
count = 0
for commit in commits_more['date']:
    commit = commit[:10]
    commit = datetime.strptime(commit, '%Y-%m-%d')
    two_weeks_ago = datetime.now()-timedelta(days = 14)
    if commit > two_weeks_ago:
        count +=1
```

```python
no_of_commits_past_2_weeks = count
no_of_commits_past_2_weeks

```
